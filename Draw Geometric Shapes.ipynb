```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Create a blank image
img = np.zeros((500, 500, 3), dtype='uint8')

# Draw the bus body (a rectangle)
bus_color = (0, 0, 255)  # Red color in BGR format
cv2.rectangle(img, (100, 200), (400, 300), bus_color, -1)  # Fill the rectangle

# Draw the bus windows (rectangles on the bus body)
window_color = (255, 255, 255)  # White color in BGR format
cv2.rectangle(img, (150, 220), (250, 260), window_color, -1)  # Window 1
cv2.rectangle(img, (275, 220), (375, 260), window_color, -1)  # Window 2

# Draw the wheels of the bus (two circles)
wheel_color = (0, 0, 0)  # Black color in BGR format
cv2.circle(img, (150, 350), 30, wheel_color, -1)  # Wheel 1
cv2.circle(img, (350, 350), 30, wheel_color, -1)  # Wheel 2

# Add text to label the bus
font = cv2.FONT_HERSHEY_SIMPLEX
text_color = (0, 255, 0)  # Green color in BGR format
cv2.putText(img, 'School Bus', (180, 450), font, 1, text_color, 2, cv2.LINE_AA)

# Display the image using matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
plt.axis('off')  # Turn off axis labels
plt.show()
```

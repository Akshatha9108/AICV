# DETECT FACES

```python
import cv2

# Try to read the image
img = cv2.imread('cricketers.jpg')

# Check if the image was successfully loaded

# Convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the required Haar cascade XML classifier
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Apply the face detection method on the grayscale image
faces_rect = haar_cascade.detectMultiScale(gray_img, 1.1, 9)

# Iterating through rectangles of detected faces
for (x, y, w, h) in faces_rect:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with detected faces
cv2.imshow('Detected faces', img)

# Wait for a key press and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# DETECT EYES

```python
import cv2

# Try to read the image
img = cv2.imread('eyes.webp')

# Convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the required Haar cascade XML classifier
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Apply the face detection method on the grayscale image
eyes_rect = eye_cascade.detectMultiScale(gray_img, 1.1, 9)

# Iterating through rectangles of detected faces
for (x, y, w, h) in eyes_rect:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with detected faces
cv2.imshow('Detected eyes', img)

# Wait for a key press and then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# DETECT VIDEOS

```python
import cv2
cap = cv2.VideoCapture('virat.mp4')
# Check if camera opened successfully
if (cap.isOpened()== False):
    print("Error opening video file")
# Read until video is completed
while(cap.isOpened()):
# Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
    # Display the resulting frame
        gray_img = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
        face_rect = haar_cascade.detectMultiScale(gray_img,1.1,9)
        for(x,y,w,h) in face_rect:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        cv2.imshow('Detected faces',frame)
    # Press Q on keyboard to exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    # Break the loop
    else:
        break
# When everything done, release
# the video capture object
cap.release()
cv2.destroyAllWindows()
```
